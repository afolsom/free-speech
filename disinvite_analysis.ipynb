```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("disinvitationattempts.csv")
df.head()
```

```python
df2=df[["Year","Disinvitation?","From Right or Left of Speaker"]]
df2.head()
```

```python
df2=df2.dropna().reset_index(drop=True)
```

```python
df2_dis=df2[df2["Disinvitation?"]=="Yes"].reset_index(drop=True)
df2_dis.head()
```

```python
df2_nodis=df2[df2["Disinvitation?"]=="No"].reset_index(drop=True)
df2_nodis.head()
```

```python
import seaborn as sns
```

```python
df2["From Right or Left of Speaker"].value_counts()
```

```python
df2_dis["From Right or Left of Speaker"].value_counts()
```

```python
df2_nodis["From Right or Left of Speaker"].value_counts()
```

```python
df2_year=df2[["Year","From Right or Left of Speaker"]]
```

```python
sns.catplot(x="From Right or Left of Speaker", y="Year", kind="swarm", data=df2_year)
```

```python
sns.catplot(y="Year", x="From Right or Left of Speaker", kind="box", data=df2_year)
```

```python
sns.catplot(x="From Right or Left of Speaker", y="Year", hue="Disinvitation?",kind="box", data=df2)
```

```python
sns.catplot(x="From Right or Left of Speaker", y="Year", kind="boxen", data=df2)
```

```python
sns.catplot(x="From Right or Left of Speaker", y="Year", hue="Disinvitation?", kind="violin", split=True, data=df2)
```
